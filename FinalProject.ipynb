```python
df.isnull()
```

```python
#droping Unamed unescessery columns
df.drop('Character Comments', inplace=True, axis=1)
df.drop('Story Comments', inplace=True, axis=1)
df.drop('Setting Comments', inplace=True, axis=1)
df.drop('Music Comments', inplace=True, axis=1)
df.drop('Extra Content', inplace=True, axis=1)
df.drop('Miscellaneous Comments', inplace=True, axis=1)

df.drop('Unnamed: 27', inplace=True, axis=1)
df.drop('Unnamed: 28', inplace=True, axis=1)
df.drop('Unnamed: 29', inplace=True, axis=1)
```

```python
df.head()
```

```python
miss = df.isnull()
miss.sum()
#CHecking amount of missing values in each row
```

```python
#Replacing numarical missing values with mean, median and others with most accoured
A = df["My Score"].astype("float").mean(axis=0)
df["My Score"].replace(np.nan,A,inplace = True)

A = df["MAL Score"].astype("float").mean(axis=0)
df["MAL Score"].replace(np.nan,A,inplace = True)

A = df["Weighted Score Diff"].astype("float").mean(axis=0)
df["Weighted Score Diff"].replace(np.nan,A,inplace = True)

A = df["Score Diff"].astype("float").mean(axis=0)
df["Score Diff"].replace(np.nan,A,inplace = True)
```

```python
#replacing string values with most accurred
M = df["MAL Count"].value_counts().idxmax()
df["MAL Count"].replace(np.nan, M, inplace=True)

M = df["Under or Over"].value_counts().idxmax()
df["Under or Over"].replace(np.nan, M, inplace=True)

M = df["MAL Count"].value_counts().idxmax()
df["MAL Count"].replace(np.nan, M, inplace=True)



#dropping rest of the rows with any missing values
df = df.dropna(axis = 0, how ='any') 
miss = df.isnull()
print("No missing values remains anymore")
miss.sum()
```
